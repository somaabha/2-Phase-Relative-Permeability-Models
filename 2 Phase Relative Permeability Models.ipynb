# 2 Phase Relative Permeability Models

## Corey's 2 Phase Model - Valid for Drainage in a Consolidated Rock.

Swn=(Sw-Swi)/(1-Swi)

Krw=Swn^4

Kro=(1-Swn)^2 * (1-Swn^2)

```python
#importing libraries 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
Sw = np.linspace(0,1,100)
Swi = 0.20
Swn=(Sw-Swi)/(1-Swi)
Krw=Swn**4
Kro=((1-Swn)**2) * ((1-Swn**2))
```

```python
plt.plot(Sw,Krw,Sw,Kro)
plt.title("Rel-Perm by Corey's Model")
plt.xlabel('Sw')
plt.ylabel('Krel(o,w)')
plt.grid()
plt.legend(['Krel-water(Wetting P.)','Krel-Oil(Non-Wetting P.)'])
```

## Naar-Henderson's Model

Swn=(Sw-Swi)/(1-Swi)

Krw=Swn^4

Kro=[(1-2*Swn)^1.5] *[2- ((1-2*Swn)0.5)]

```python
Sw = np.linspace(0,1,100)
Swi = 0.20
Swn=(Sw-Swi)/(1-Swi)
Krw=Swn**4
Kro=((1-2*Swn)**(1.5)) * (2- ((1-2*Swn)**0.5))
```

```python
plt.plot(Sw,Krw,label='Krel-water(Wetting P.)')
plt.plot(Sw,Kro,label='Krel-Oil(Non Wetting P.)')
plt.title("Rel-Perm by Naar-Henderson's Model")
plt.xlabel('Sw')
plt.ylabel('Krel(o,w)')
plt.grid()
plt.legend(['Krel-water(Wetting P.)','Krel-Oil(Non-Wetting P.)'])
```
